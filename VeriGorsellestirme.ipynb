```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
```

```python
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')

plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
plt.legend(('sin', 'cos'))
```

```python
fig = plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x), '-')

plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
plt.legend(('sin', 'cos'))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), '--', linewidth=4, color='red')

plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
plt.legend(('sin', 'cos'))
```

```python
fig.savefig('data.png')
```

```python
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['petal_length'].plot()
```

```python
sns.lineplot(data=iris, y='sepal_length', x=iris.index, hue='species')
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
ax.scatter(iris['sepal_length'], iris['sepal_width'])
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
data = iris.groupby(by ='species').mean()['sepal_length']
ax = sns.barplot(x=data.index, y=data.values)
ax.set_title('Sepal Length Means')
ax.set_ylabel('Max')

```

```python
sns.histplot(data = iris, x='sepal_width')
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr(), annot=True)
```

```python
sns.boxplot(data=iris, y='sepal_length', x='species')
```
